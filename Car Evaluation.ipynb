https://archive.ics.uci.edu/ml/datasets/Car+Evaluation

## Goal
Build a model with KNeighborsClassifier.
#### Highest Accuracy: 97,10%

## Procedure

```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model, preprocessing

#import data
df = pd.read_csv('input/car.data')
```

```python
df.head()
```

```python
display(df.columns, df.dtypes, df.shape, df.describe())
```

```python
#change data to numerical
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(df['buying']))
maint = le.fit_transform(list(df['maint']))
door = le.fit_transform(list(df['door']))
persons = le.fit_transform(list(df['persons']))
lug_boot = le.fit_transform(list(df['lug_boot']))
safety = le.fit_transform(list(df['safety']))
cls = le.fit_transform(list(df['class']))
```

```python
#target
predict = 'class'

#split data
X = list(zip(buying, maint, door, persons, lug_boot, safety))
y = list(cls)

#train the modell
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1)
model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
predicted = model.predict(x_test)
print(acc)
```

- model with up to 97.10% accuracy (test size = 10%, n_neigbors = 7)

```python
#compare the prediction with actual values
col = ['unacc', 'acc', 'good', 'vgood']
for x in range(len(x_test)):
    print('Predicted: ', col[predicted[x]], 'Data: ', x_test[x], 'Actual: ', col[y_test[x]])    
```
